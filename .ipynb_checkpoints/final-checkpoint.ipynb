El objetivo de la siguiente Notebook es mostrar como se lleva a cabo una corrección de color. Para ello, se dispondrá de una imagen base o de partida, a la que se denominará 'Original'. La misma será modificada, simulando una distorsión/modificación/perturbación en la adquisición de una imagen; denominada 'Modificada'. A esta última, se le aplicará la corrección, intentando conseguir nuevamente la imagen 'Original', haciendo uso de un cuadro de colores de 'referencia'. El propósito del mismo es conocer a priori los colores (conocer sus valores digitales), a modo de poder medir qué tanto han cambiado y de qué forma. En la práctica, el cuadro de referencia sufre las mismas modificaciones que la escena de interés, de modo que ambos sufren los mismos cambios.

Ahora se aplica la corrección de color. El método, también explicado en el código, se compone de los siguientes pasos:

- Se obtienen las curvas que se adaptan a cada canal con los tonos grises centrales. Ello sirve para balancear los blancos.
- Se obtiene una Matriz de Corrección de Color (CCM) de la siguiente manera:

        - Ro = a0 + a11*R + a12*G + a13*B + a21*R² + a22*G² + a23*B²
        - Go = b0 + b11*R + b12*G + b13*B + b21*R² + b22*G² + b23*B²
        - Bo = c0 + c11*R + c12*G + c13*B + c21*R² + c22*G² + c23*B²

        donde:
        - Ro,Go,Bo son la salida de un pixel.
        - R,G,B son la entrada y los colores input de un pixel.
        - a0,b0,c0 son valores offset/bias de transformación.
        - aij,bij,cij son los coeficientes de transformación de la CCM. Aquellos coeficientes x2j corresponden a una relación cuadrática y aquellos x1j a una relación lineal. Estos coeficientes se obtienen tras ajustar una curva de transformación para cada canal siguiendo un criterio LS (Least Squares) a partir de los colores del patrón Observado y los de Referencia. Se podrían haber utilizado coeficientes x3j tal que e.g: a31*R*G, pero en la práctica no traía beneficios adicionales para el caso de estudio en cuestión (el de las semillas) y el algoritmo se tornaba más lento.
        
- Se revierte la compresión gamma existente.
- Se le aplica el arreglo de curvas.
- Se sustrae la media de la imagen, se aplica la CCM y se vuelve a añadir la media.
- Nuevamente se le aplica la compresión gamma.

A continuación se aplica la corrección y se visualiza el resultado:

```python
# apply color correction
color_corrector = TransformBased(Nh=10, Nv=1)
img_corrected = color_corrector._correct(modified, centers_modified, centers)
```

```python
corrected = cv.cvtColor(img_corrected, cv.COLOR_BGR2RGB)
plt.figure()
plt.imshow(corrected)
```

Se busca que el error obtenido sea menor que en el caso de prescindir del algoritmo.

```python
def color_error(reference, target):
    error = np.square(np.subtract(reference, target))
    error = error.reshape((error.shape[0] * error.shape[1], error.shape[2]))
    return np.array([(x[0] + x[1] + x[2]) for x in error]).mean()

# get error
original = original.astype(np.float32)
modified = modified.astype(np.float32)
corrected = img_corrected.astype(np.float32)
error1 = color_error(original, modified)
error2 = color_error(original, corrected)
```

Se corrobora que sea así

```python
print(error1)
print(error2)
```

El error obtenido puede ser reducido de las siguientes maneras:

- Utilizando una LUT en 3 dimensiones de 256x256x256 de manera de disponer de la conversión de todos los colores existentes. Los problemas son: la cantidad de memoria disponible, ya que el código debe ser ejecutado en la nube, y la imposibilidad de albergar esa cantidad de colores en la aplicación real.
- Utilizando más coeficientes en la matriz CCM y/o en la curvas de canal. El problema es que el tiempo de procesamiento adicional no justifica la baja mejoría obtenida. Se recuerda, que este proceso será atravesado por muchas imágenes: ello demanda un mayor tiempo de procesamiento y recursos en la nube, que se traduce en un costo más elevado.
- Utilizando algoritmos más complejos. Existen redes neuronales que se ocupan de estas tareas, pero entrenarlas demanda muchos recursos (búsqueda de arquitectura óptima, digitalización para generar un dataset, etc.)
- Realizar un análisis más exhaustivo de las imágenes a analizar, modelar el sistema físico de digitalización a modo de mejorar propiedades ópticas del escenario para la obtención de la imagen y cancelar efectos no deseados.
